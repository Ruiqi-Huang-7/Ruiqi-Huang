```python
# Import libraries
# 导入库（W1内容）

import pandas as pd        # For handling data  用于数据处理
import numpy as np         # For numerical analysis  用于数值计算
import matplotlib.pyplot as plt   # For plotting  用于绘图
```

```python
# Load the dataset
# 读取数据文件

data = pd.read_csv("project_1.csv")
```

```python
# Convert time column to datetime
# 将时间列转换为日期格式

data["time"] = pd.to_datetime(data["time"])
```

```python
# Check summary statistics
# 查看描述统计

print(data[["FSNS", "TREFHT"]].describe())
```

```python
# Remove missing values
# 删除缺失值

data = data.dropna(subset=["FSNS", "TREFHT"])
```

```python
# Plot temperature over time
# 绘制温度时间序列

plt.figure()

plt.plot(data["time"], data["TREFHT"])

plt.xlabel("Time")
plt.ylabel("TREFHT (K)")
plt.title("Time Series of Near-Surface Temperature")

plt.show()
```

```python
# Plot radiation over time
# 绘制短波辐射时间序列

plt.figure()

plt.plot(data["time"], data["FSNS"])

plt.xlabel("Time")
plt.ylabel("FSNS")
plt.title("Time Series of Net Shortwave Radiation")

plt.show()
```

```python
# Scatter plot of FSNS vs Temperature
# 绘制散点图

plt.figure()

plt.scatter(data["FSNS"], data["TREFHT"])

plt.xlabel("FSNS")
plt.ylabel("TREFHT (K)")
plt.title("Relationship Between Radiation and Temperature")
```

```python
# Linear regression using polyfit
# 使用 polyfit 进行线性回归

m, b = np.polyfit(data["FSNS"], data["TREFHT"], 1)

plt.plot(data["FSNS"], m*data["FSNS"] + b)

plt.show()
```

```python
# Calculate correlation coefficient
# 计算皮尔逊相关系数

r = np.corrcoef(data["FSNS"], data["TREFHT"])[0,1]

print("Correlation coefficient:", r)
```

```python
# Calculate R squared
# 计算决定系数

r_squared = r**2

print("R squared:", r_squared)
```
